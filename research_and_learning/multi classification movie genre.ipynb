```python
# Text Gathering

# from google drive
data_path = "/content/drive/MyDrive/Datasets/movie-genre-prediction/train.csv"
```

```python
import pandas as pd
import tensorflow as tf
```

```python
# load the dataset and create a pipeline with tf.data

class LoadAndPreprocess():

    def __init__(self, data_path,):
        super().__init__()
        self.data_path = data_path
        self.vector_layer = tf.keras.layers.TextVectorization()


    def load(self):
        dataset = pd.read_csv(self.data_path)
        return dataset

    def preprocess(self, dataset, feature_col, label_col):
        input = dataset[feature_col]
        output = dataset[label_col]

        return input, output

    def vectorize(self, dataset, feature_col, label_col):
        for col in [feature_col, label_col]:
            self.vector_layer.adapt(dataset[col])

        print(self.vector_layer.vocabulary_size())
        print(self.vector_layer.get_vocabulary())
```

```python
lap = LoadAndPreprocess(data_path)
dataset = lap.load()
print(dataset.value_counts("genre"))
```
