```python
import tensorflow as tf
import os
import pandas as pd
```

```python
source = "https://ai.stanford.edu/~amaas/data/sentiment/aclImdb_v1.tar.gz" 

imdb = tf.keras.utils.get_file(fname = "aclImdb_v1", origin=source, untar=True, cache_dir=".", cache_subdir='')
```

```python
imdb
```

```python
dataset = os.path.join(os.path.dirname(imdb), 'aclImdb')
dataset
```

```python
os.listdir(dataset)
```

```python
train_ds = os.path.join(dataset, 'train')
os.listdir(train_ds)
```

```python
train_neg_ds = os.path.join(train_ds, 'neg')
train_pos_ds = os.path.join(train_ds, 'pos')
```

```python
file_ = open(f"{train_pos_ds}/8695_9.txt")
```

```python
df = pd.DataFrame(columns=["review", "label"])
```

```python
neg_files = os.listdir(train_neg_ds)
for nfile in neg_files:
    nread = open(f"{train_neg_ds}/{nfile}")
    text = nread.readlines()[0]
    df2 = pd.DataFrame({"review":text, "label":0})
    df.concat([df2],ignore_index = True)
    del df2
    nread.close()
print('DONE')
```
