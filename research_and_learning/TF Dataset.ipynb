```python
import tensorflow as tf
```

printing the data instead of printing entire elements

```python
dataset = tf.data.Dataset.range(10)
dataset
```

```python
for info in dataset.as_numpy_iterator():
    print(info)
```

if you wanna apply any functionality on dataset use -> apply

```python
def even(ds):
    return ds.filter(lambda info: info>=5)

for info in dataset.apply(even).as_numpy_iterator():
    print(info)
```

Batch

```python
for info in dataset.batch(3,True):
    print(info)
```

Concatenating

```python
foo = tf.data.Dataset.from_tensors("foo").repeat()
bar = tf.data.Dataset.from_tensors("bar").repeat()
baz = tf.data.Dataset.from_tensors("baz").repeat()

choice = tf.data.Dataset.range(3).repeat(3)


dataset = tf.data.Dataset.choose_from_datasets([foo,bar,baz],choice)
for info in dataset.as_numpy_iterator():
    print(info)
```

```python
ds1 = tf.data.Dataset.range(3)
ds2 = tf.data.Dataset.range(3,7)
ds = ds1.concatenate(ds2)
for info in ds.as_numpy_iterator():
    print(info)
```

```python
from pathlib import Path

SO_FILE = Path(so_file).parent
print(SO_FILE)
```

```python
train_ds = SO_FILE/'train'
print(train_ds)
```

```python
dataset = tf.keras.utils.text_dataset_from_directory(train_ds)
```

```python
vector_layer = tf.keras.layers.TextVectorization(output_mode='int')

data = dataset.map(lambda text, labels: text)
ds = vector_layer.adapt(data)
```
