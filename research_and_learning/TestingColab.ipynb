```
x_even,x_odd = x[0],x[1]
```

```
x_even
```

```
for i in x_even+x_odd:
  print(i)
```

```
x_final=[]
for i in x_even+x_odd:
  x_final.append(i)
```

```
x_sort = sorted(x_final, key=lambda a:a.split(" ")[-1])
```

```
x_sort
```

```
l = len(x_sort)
for i in x_sort:
  # print(type(i.split(" ")[-1]))
  # print(l-1)
  if int(i.split(" ")[-1])==l-1:
    print(i)
  else:
    print(i,end=",\n")
```
