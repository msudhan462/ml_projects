```python
from tensorflow.keras import utils
```

```python
data_dir = utils.get_file(origin="https://storage.googleapis.com/download.tensorflow.org/data/stack_overflow_16k.tar.gz", untar=True, cache_dir='stack_overflow', cache_subdir="")
```

```python
from pathlib import Path
```

```python
data_dir = Path(data_dir)
```

```python
data_dir
```

```python
data_dir = data_dir.parent
```

```python
list(data_dir.parent.iterdir())
```

```python
dataset_train = data_dir/"train"
```

```python
# for sub in list(dataset_train.iterdir()):
#     print(list(sub.iterdir()))
```

```python
train_raw = utils.text_dataset_from_directory(dataset_train, batch_size=32, seed=40, validation_split=0.2, subset='training')
```

```python
for data, labels in train_raw.take(1):
    for _ in range(5):
        print(data.numpy()[_], labels.numpy()[_])
```
