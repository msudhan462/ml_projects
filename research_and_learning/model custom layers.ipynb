```python
# non trainable arguments
class Linear(tf.keras.layers.Layer):

    def __init__(self, units=32, input_dim=32):
        super(Linear, self).__init__()
        self.w = self.add_weight(
            shape = (input_dim, units), initializer = "random_normal", trainable = False
        )
        self.b = self.add_weight(
            shape = (units, ), initializer = "zeros", trainable = False
        )

    def call(self, inputs):
        return tf.reduce_sum(inputs, axis=1)
        # return tf.matmul(inputs, self.w) + self.b
    
inputs = tf.Variable(
    [
        [1, 2],
        [3, 4],
        [5, 6]
    ],
    dtype = "float"
)

print(inputs)
input_dim = inputs.shape[1]
linear = Linear(4,input_dim)
print(linear)
print(linear(inputs))
linear.trainable_variables
```

```python
# trainable parameters should be in build() method

class UsingBuild(tf.keras.layers.Layer):

    def __init__(self, units, name=None):
        super(UsingBuild, self).__init__(name=name)
        self.units = units
    
    def build(self, input_shape):
        print(f"input_shape is {input_shape}")
        self.w = self.add_weight(
            shape = (input_shape[-1], self.units), initializer = "random_normal", trainable = True, name="weights"
        )
        self.b = self.add_weight(
            shape = (self.units, ), initializer = "zeros", trainable = True, name="bais"
        )

    def call(self, inputs):
        return tf.matmul(inputs, self.w) + self.b
    

inputs = tf.Variable(
    [
        [0, 1, 3, 4, 5],
        [2, 3, 4, 5, 7],
        [3, 4, 9, 8, 7]
    ],
    dtype = 'float'
)
ub = UsingBuild(32)
print(ub(inputs))
```

```python
for var in ub.trainable_variables:
    print(f"{var.name} ==> {var.shape}")
```

```python
# we can use Layer in another layer

class MultiLayerNetwork(tf.keras.layers.Layer):

    def __init__(self):
        super(MultiLayerNetwork, self).__init__()
        self.linear1 =UsingBuild(32, "linear1")
        self.linear2 =UsingBuild(32, "linear2")
        self.linear3 =UsingBuild(1, "linear3")
    
    def call(self, inputs):
        x1 = self.linear1(inputs)
        x1 = tf.nn.relu(x1)

        x2 = self.linear2(x1)
        x2 = tf.nn.relu(x2)

        x3 = self.linear3(x2)
        return tf.nn.softmax(x3)

mln = MultiLayerNetwork()
print("weights", len(mln.weights))
print(mln(inputs))
print("weights", len(mln.weights))
```
