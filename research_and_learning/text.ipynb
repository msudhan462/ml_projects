```python
import numpy as np
from pprint import pprint
```

```python
samples = [
    "the cat sat on the mat",
    "the dog ate my food on the table."
]

token_dict = {}
for sample in samples:
    for word in sample.split(" "):
        if word not in token_dict:
            token_dict[word] = len(token_dict) + 1
print(token_dict)
```

```python
max_length = 10
results = np.zeros(
    shape=(len(samples), max_length, max(token_dict.values())+1)
)
results.shape
```

```python
for index, sample in enumerate(samples):
    for index2, word in enumerate(sample.split(" ")):
        value = token_dict.get(word)
        results[index,index2,value] = 1

```

```python
## character level encoding


samples = [
    "the cat sat on the mat",
    "the dog ate my food on the table."
]

char_di = {}

for sample in samples:
    for ch in sample:
        if ch not in char_di:
            char_di[ch] = max(char_di.values())+1
```
